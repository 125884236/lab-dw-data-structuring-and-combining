# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
import pandas as pd

df_marketing = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")

pd.set_option("display.max_columns",None)

df_marketing.drop(["unnamed:_0"], axis = 1, inplace = True)

df_marketing["total_claim_amount"] = df_marketing["total_claim_amount"].round(2)

df_marketing["number_of_open_complaints"] = df_marketing["number_of_open_complaints"].astype(int)

df_marketing["months_since_last_claim"] = df_marketing["months_since_last_claim"].astype(int)

df_marketing.rename(columns = {"employmentstatus" : "employment_status"}, inplace=True)

df_marketing["customer_lifetime_value"] = df_marketing["customer_lifetime_value"].round(2)

renew_offer_new_values = {
    "Offer1" : "Offer 1",
    "Offer2" : "Offer 2",
    "Offer3" : "Offer 3",
    "Offer4" : "Offer 4"
}

df_marketing["renew_offer_type"] = df_marketing["renew_offer_type"].replace(renew_offer_new_values)
```

```python
df_marketing.head(1)
```

```python
# You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue.
# Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).

df_marketing.groupby("sales_channel").agg({"customer": "count"})
```

```python
df_marketing[["sales_channel", "coverage"]].groupby("sales_channel").count()
```

```python
# Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

# Las mujeres con menor nivel educativo son las que más tiempo permanecen como clientas de la aseguradora, seguidas de las que tienen un máster.
# En el caso de los hombres se trata de los que tienen un máster, seguido de los que tienen tan solo la enseñanza obligatoria.
# Para ambos sexos, los clientes que menos tiempo permanecen son los que tienen un doctorado.

# No obstante, los datos son muy similares tanto para hombres como para mujeres para todos los niveles educativos.

df_marketing.pivot_table(index = "gender", columns = "education", values=["customer_lifetime_value"], aggfunc = "mean")
```
